```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
 return 1/(1 + np.exp(-x))

def min_max_normalization(x):
    return (x - np.min(x)) / (np.max(x) - np.min(x))

def standardization(x):
    return (x - np.mean(x)) / np.std(x)

def sum_normalization(x):
    return x / np.sum(x)
```

```python
# Sample data
x = np.random.normal(0, 1, 1000)  # Sample 1000 points from a normal distribution
min_x, max_x = np.min(x), np.max(x)

# Min-Max normalization function
x_min_max = min_max_normalization(x)

# Sigmoid normalization function
x_sigmoid = sigmoid(x)

# Standardization function
x_standard = standardization(x)

# Sum normalization function
x_sum = sum_normalization(x)

# Plot
plt.scatter(x, x_min_max, alpha=0.6, label='Min-Max Normalization')
plt.scatter(x, x_sigmoid, alpha=0.6, label='Sigmoid Normalization')
plt.xlabel('Sample Index')
plt.ylabel('Normalized Value')
plt.legend()
```
